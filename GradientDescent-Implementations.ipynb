```python
# Finding gradient descent with Python
import numpy as np
import matplotlib.pyplot as plt

# Generate random input features (100 samples, 3 features)
x = np.random.uniform(low=-10, high=10, size=(100, 1))
y = np.random.uniform(low=-10, high=10, size=(100, 1))
z = np.random.uniform(low=-10, high=10, size=(100, 1))
inputs = np.column_stack((x, y, z))  # Combine features into input matrix
print(inputs.shape)  # Display shape of input matrix (100, 3)

# Create target values with a linear relationship to inputs plus some noise
noise = np.random.uniform(low=-1, high=1, size=(100, 1))
target = 150*x + 20000*y + 0.0009*z + noise  # Linear function with coefficients
print(target.shape)  # Display shape of target matrix (100, 1)

# Initialize random weights and bias for the model
weights = np.random.uniform(low=-1, high=1, size=(3, 1))
print(weights.shape)  # Display shape of weights matrix (3, 1)
bias = np.random.uniform(low=-1, high=1, size=(1, 1))

# Set hyperparameters for gradient descent
rate = 0.00001  # Learning rate
losses = []  # List to store loss values
epoch = []  # List to store epoch numbers

# Gradient descent training loop
for i in range(10):
    # Forward pass: compute predictions
    output = np.dot(inputs, weights) + bias
    print(output.shape)  # Display shape of output (100, 1)
    
    # Compute error
    delta = output - target
    
    # Compute mean squared error loss
    loss = np.sum((delta**2)/2)
    
    # Update weights and bias using gradient descent
    weights = weights - rate * np.dot(inputs.T, delta)  # Update weights
    bias = bias - rate * np.sum(delta)  # Update bias
    
    # Store loss and epoch for plotting
    losses.append(loss)
    epoch.append(i)
    print(loss)  # Print current loss

# Print final model parameters and training history
print(weights)
print(bias)
print(losses)
print(epoch)

# Plot training curve: epoch vs losses
plt.plot(epoch, losses)
plt.scatter(epoch, losses)
plt.xlabel("epoch")
plt.ylabel("losses")
plt.show()

# Plot model performance: actual target vs predicted target
predicted = np.dot(inputs, weights) + bias
plt.plot(target, predicted)
plt.scatter(target, predicted)
plt.xlabel("target")
plt.ylabel("predicted")
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Generate random input data (x and y coordinates)
x = np.random.uniform(low=10, high=-10, size=(100, 1))  # 100 random x values between -10 and 10
y = np.random.uniform(low=10, high=-10, size=(100, 1))  # 100 random y values between -10 and 10

# Combine x and y into a single input array
inputs = np.column_stack((x, y))

# Create target values using a linear equation: 15x + 25y + 2
target = 15*x + 25*y + 2

# Define output size for the model
outputsize = 1

# Save the generated data to a file
np.savez("tf2", inp=inputs, tar=target)

# Load the saved data
training_data = np.load("tf2.npz")

# Create a simple neural network model with one dense layer
# Initialize weights and biases with small random values between -0.1 and 0.1
model = tf.keras.Sequential(tf.keras.layers.Dense(outputsize, 
                                                kernel_initializer=tf.keras.initializers.RandomUniform(minval=-0.1, maxval=0.1),
                                                bias_initializer=tf.keras.initializers.RandomUniform(minval=-0.1, maxval=0.1)))

# Set up the Adam optimizer
custom_optimizer = tf.keras.optimizers.legacy.Adam()

# Compile the model with mean squared error loss function
model.compile(custom_optimizer, loss='mean_squared_error')

# Train the model for 100 epochs (first training run)
model.fit(training_data['inp'], training_data['tar'], epochs=100, verbose=2)

# Train the model again and store the training history
history = model.fit(training_data['inp'], training_data['tar'], epochs=100, verbose=2)

# Plot the loss over epochs
# plt.scatter(epochs,loss)  # This line is commented out as 'epochs' and 'loss' are not defined
plt.plot(history.history['loss'])
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Generate random input features (x, y, z, q) with values between -10 and 10
x = np.random.uniform(low=-10, high=10, size=(100, 1))
y = np.random.uniform(low=-10, high=10, size=(100, 1))
z = np.random.uniform(low=-10, high=10, size=(100, 1))
q = np.random.uniform(low=-10, high=10, size=(100, 1))

# Generate random noise for the target variable
noise = np.random.uniform(low=-10, high=10, size=(100, 1))

# Combine all input features into a single array
input = np.column_stack((x, y, z, q))

# Create target variable using a linear equation: 4x + 3y + 2z + q + 10 + noise
target = 4*x + 3*y + 2*z + 1*q + 10 + noise

# Save input and target data to a compressed numpy file
np.savez("new", inp=input, tar=target)

# Load the saved data
loki = np.load("new.npz")

# Define the output size for the model
outputsize = 1

# Create a simple linear regression model with one Dense layer
model = tf.keras.Sequential((tf.keras.layers.Dense(outputsize, 
                                                 kernel_initializer=tf.keras.initializers.RandomUniform(minval=-0.1, maxval=0.1),
                                                 bias_initializer=tf.keras.initializers.RandomUniform(minval=-0.1, maxval=0.1))))

# Define Adam optimizer
custom_optimizer = tf.keras.optimizers.Adam()

# Compile the model with mean squared error loss
model.compile(custom_optimizer, loss="mean_squared_error")

# Train the model for 100 epochs
model.fit(loki["inp"], loki["tar"], epochs=100, verbose=2)

# Get and print the weights and bias after initial training
w1, b1 = model.layers[0].get_weights()
print(f'Before Weights and bias{w1, b1}')

# Train the model for another 100 epochs and store the training history
history = model.fit(loki["inp"], loki["tar"], epochs=100, verbose=2)

# Get and print the weights and bias after additional training
w2, b2 = model.layers[0].get_weights()
print(f'after Weights and bias{w2, b2}')

# Plot the loss over epochs
plt.plot(history.history["loss"])
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate random feature data (100 samples, 4 features)
x = np.random.uniform(low=-10, high=10, size=(100, 1))  # Feature 1
y = np.random.uniform(low=-10, high=10, size=(100, 1))  # Feature 2
z = np.random.uniform(low=-10, high=10, size=(100, 1))  # Feature 3
q = np.random.uniform(low=-10, high=10, size=(100, 1))  # Feature 4
noise = np.random.uniform(low=-10, high=10, size=(100, 1))  # Random noise

# Create target variable with known coefficients (4, 3, 2, 1) and bias (10)
target = 4*x + 3*y + 2*z + 1*q + 10 + noise

# Combine all features into a single input matrix
a = np.column_stack((x, y, z, q))

# Initialize model parameters randomly
weights = np.random.uniform(low=-1, high=1, size=(4, 1))  # Initial weights
bias = np.random.uniform(low=-1, high=1, size=(1, 1))     # Initial bias

# Set hyperparameters
rate = 0.001  # Learning rate
n = len(a)    # Number of samples

# Lists to track training progress
loss1 = []
epoch = []

# Training loop - implementing gradient descent
for i in range(100):
    # Forward pass: compute predictions
    output = np.dot(a, weights) + bias
    
    # Compute error
    delta = output - target               # prediction error
    
    # Compute mean squared error loss
    loss = np.mean(delta**2) / 2
    
    # Store loss and epoch for plotting
    loss1.append(loss)
    epoch.append(i)
    
    # Compute gradients
    grad_w = np.dot(a.T, delta) / n      # averaged gradient for weights
    grad_b = np.sum(delta) / n           # averaged gradient for bias
    
    # Update parameters using gradient descent
    weights = weights - rate * grad_w
    bias = bias - rate * grad_b

# Plot the training loss over epochs
plt.plot(epoch, loss1, label="loss vs epoch")
plt.xlabel("epoch")
plt.ylabel("loss1")
plt.legend()
plt.show()

# Compare actual target values with model predictions
predicted = np.dot(a, weights) + bias
plt.plot(target, c="blue", label="real")
plt.plot(predicted, c="red", label="predicted")
plt.legend()
plt.show()
```
